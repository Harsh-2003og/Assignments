Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
# To create a scatter plot.
plt.scatter(x,y)
plt.xlabel("This is the x-Axis")
plt.ylabel("This is the y-Axis")
plt.title("Example of scatter plot")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

The subplot() function is used to create multiple plots within the same figure. It allows us to arrange multiple plots in a grid of rows and columns, and specify the location of each plot using its row and column index.

```python
# For Line - 1
x1 = np.array([6, 1, 2, 3, 4, 5]) 
y1 = np.array([8, 100, 200, 300, 400, 500])

# For Line -2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([0, 50, 100, 150, 200, 250])

# For Line - 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([0, 10, 20, 30, 40, 50])

# For Line - 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([0, 5, 10, 15, 20, 25])

# Create a 2x2 grid of subplots and plot each line on a separate subplot
fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles to the subplots
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Display the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,5))
plt.bar(company, profit)
plt.xlabel("This is the categorical data")
plt.ylabel("This is the numeric Data")
plt.title("This is the bar graph")
```

```python
# To Create a horizontal bar graph (Just put barh).
plt.figure(figsize=(6,5))
plt.barh(company, profit)
plt.xlabel("This is the Numeric data")
plt.ylabel("This is the categorical Data")
plt.title("This is the Bar Graph")
```
